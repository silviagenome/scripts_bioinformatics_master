```python
import csv

f = open('lncrna_mus.ttl', 'w')

with open('triplets_template.ttl') as f_template:
    template = f_template.read()
    
with open('mus_data.csv') as mus_data:
    csv_mus = csv.DictReader(mus_data)
    
    index = 1
    for row in csv_mus:
        output = template.replace('{Title}', row['Title']).replace('{Year}',str(row['Year'])).replace('{index}',str(index)).replace('{PMID}',str(row['PMID'])).replace('{Expression}',row['Expression']).replace('{ID}',row['ID']).replace('{Target}',row['Target']).replace('{Pathway}',row['Pathway']).replace('{Function}',row['Function']).replace('{Species}',row['Species']).replace('{symbol}',row['symbol']).replace('{Disease}',row['Disease']).replace('{Tissue}',row['Tissue'])
        print(output)
        print(output, file = f)
        index += 1
                                                                                            
```
